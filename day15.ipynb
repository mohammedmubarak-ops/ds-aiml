```python
import pandas as pd
data = {
 "Emp ID": [101, 102, 103, 104, 105, 106],
 "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"], "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"], "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
 "Experience": [2, 3, 5, 4, 1, 3]
}
df=pd.DataFrame(data)
print(df)
#HEAD
print(df.head(2))
print(df.head(-2))
 
#TAIL
print(df.tail(3))
print(df.tail(-3))
 
#SHAPE
print(df.shape)
 
#INFO
print(df.info())
 
#VERBOSE
df.info(verbose=True)
df.info(verbose=False)
 
#RENAME
df.rename(columns={"Name": "Emp_Name"}, inplace=True)
print(df)
 
#DESCRIBE
df.describe()
```

OPERATIONS ON COLUMNS

```python
import pandas as pd
url = "https://raw.githubusercontent.com/mohammedmubarak-ops/ds-aiml/main/file2.json"
df = pd.read_json(url)
# get single column data
df["name"]
# add single column
df["salary"] = df["marks"] + 100
# add single column
df["salary"] = [100,200,300,400,500]
print(df)
```
