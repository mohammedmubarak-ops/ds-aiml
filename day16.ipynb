OPERATION ON ROWS

```python
#LOC AND ILOC
import pandas as pd
d = {
    "name": ["vishal","virat","vineet"],
    "salary": [100,200,300]
}
df = pd.DataFrame(data=d)
print(df.loc[2,"name"])
print(df.iloc[2,0])

#get single row data
print(df.iloc[1])
print("\n")

#get single row using loc
print(df.loc[1])
print("\n")

#get multi rows using iloc
print(df.iloc[0:3])
print("\n")

#get multi rows using loc
print(df.iloc[0:3])
print("\n")
```
